```python
print("Welcome to the quiz!")

score = 0

question1 = "What is the capital of France?"
answer1 = "Paris"
user1 = input(question1 + " ")
if user1.strip().lower() == answer1.lower():
    print("Correct!")
    score += 1
else:
    print("Incorrect. The correct answer is " + answer1 + ".")

question2 = "What is 2 + 2?"
answer2 = "4"
user2 = input(question2 + " ")
if user2.strip() == answer2:
    print("Correct!")
    score += 1
else:
    print("Incorrect. The correct answer is " + answer2 + ".")

print("Your score is " + str(score))
```
